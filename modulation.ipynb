```python
# the deep_learning model
import tensorflow as tf
from tensorflow.keras import layers, models, Input
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau
# definition of the dense layer
input_feature= Input(shape=(7,))
x1= layers.Dense(128, kernel_initializer='he_normal')(input_feature)
x1= LeakyReLU(alpha=0.2)(x1)
x1= layers.Dense(128, kernel_initializer='he_normal')(x1)
x1= LeakyReLU(alpha=0.2)(x1)
x1= layers.Dense(64, kernel_initializer='he_normal')(x1)
x1= LeakyReLU(alpha=0.2)(x1)
x1= layers.Dense(64, kernel_initializer='he_normal')(x1)
x1= LeakyReLU(alpha=0.2)(x1)
x1= layers.Dense(64, kernel_initializer='he_normal')(x1)
x1= LeakyReLU(alpha=0.2)(x1)
x1= layers.Dense(64, kernel_initializer='he_normal')(x1)
x1= LeakyReLU(alpha=0.2)(x1)
x1= layers.Dense(64, kernel_initializer='he_normal')(x1)
x1= LeakyReLU(alpha=0.2)(x1)
# definition of the CNN
input_spectrogram = Input(shape=(32, 36, 1))
x2= layers.Conv2D(32, (3, 3), activation='relu')(input_spectrogram)
x2= layers.MaxPooling2D((2, 2))(x2)
x2= layers.Conv2D(64, (3, 3), activation='relu')(x2)
x2= layers.MaxPooling2D((2, 2))(x2)
x2= layers.Conv2D(128, (3, 3), activation='relu')(x2)
x2= layers.Flatten()(x2)
x2= layers.Dense(128, kernel_initializer='he_normal')(x2)
x2= LeakyReLU(alpha=0.2)(x2)
x2= layers.Dense(128, kernel_initializer='he_normal')(x2)
x2= LeakyReLU(alpha=0.2)(x2)
x2= layers.Dense(128, kernel_initializer='he_normal')(x2)
x2= LeakyReLU(alpha=0.2)(x2)
x2= layers.Dense(64, kernel_initializer='he_normal')(x2)
x2= LeakyReLU(alpha=0.2)(x2)
x2= layers.Dense(64, kernel_initializer='he_normal')(x2)
x2= LeakyReLU(alpha=0.2)(x2)
x2= layers.Dense(64, kernel_initializer='he_normal')(x2)
x2= LeakyReLU(alpha=0.2)(x2)
merged= layers.concatenate([x1, x2])
# definition of the output layer
x= layers.Dense(64, kernel_initializer='he_normal')(merged)
x= LeakyReLU(alpha=0.2)(x)
x= layers.Dense(64, kernel_initializer='he_normal')(x)
x= LeakyReLU(alpha=0.2)(x)
x= layers.Dense(64, kernel_initializer='he_normal')(x)
x= LeakyReLU(alpha=0.2)(x)
x= layers.Dense(64, kernel_initializer='he_normal')(x)
x= LeakyReLU(alpha=0.2)(x)
x= layers.Dense(64, kernel_initializer='he_normal')(x)
x= LeakyReLU(alpha=0.2)(x)
x= layers.Dense(32, kernel_initializer='he_normal')(x)
x= LeakyReLU(alpha=0.2)(x)
output= layers.Dense(24, activation='softmax')(x)
model= models.Model(inputs=[input_feature, input_spectrogram], outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
def z_score_normalization(signal):
  mean= np.mean(signal)
  std= np.std(signal)
  normalized_signal= (signal-mean)/std
  threshold = 3
  outliers = np.abs(normalized_signal) > threshold
  signal[outliers] = np.median(signal)
  return signal
```

```python
def moving_average(signal, window_size):
  #signal = signal.ravel()
  return np.convolve(signal, np.ones(window_size)/window_size, mode='same')
```

```python
from scipy.ndimage import gaussian_filter1d
def gaussian_filtering(signal):
  sigma = 3  # Standard deviation for the Gaussian kernel
  filtered_signal = gaussian_filter1d(signal, sigma=sigma)
  return filtered_signal
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
def extract_spectral_features(complex_data, SNR):
  #calculating PSD
  psd_values= np.zeros((complex_data.shape[0], 3))
  spectral_entropy= np.zeros((complex_data.shape[0],1))
  spectral_flatness= np.zeros((complex_data.shape[0],1))
  for i in range(complex_data.shape[0]):
    frequencies,psd= welch(complex_data[i],  nperseg=1024)
    psd_normalized = psd / np.sum(psd)
    psd_normalized += 1e-12
    spectral_entropy[i] = -np.sum(psd_normalized * np.log(psd_normalized))
    psd_values[i,0]= np.mean(psd)
    psd_values[i,1]= np.max(psd)
    psd_values[i,2]= np.trapz(psd, frequencies)
    geometric_mean = np.exp(np.mean(np.log(psd + 1e-10)))
    arithmetic_mean= np.mean(psd)
    spectral_flatness[i]= geometric_mean/arithmetic_mean
  # calculating RMS( Root Mean Square)
  rms= np.zeros((complex_data.shape[0],1))
  for i in range(complex_data.shape[0]):
    rms[i] = np.sqrt(np.mean(complex_data[i, :]**2))
  spectral_features=np.hstack((psd_values,spectral_flatness,spectral_entropy,rms))
  spectral_features= scaler.fit_transform(spectral_features)
  SNR=scaler.fit_transform(SNR)
  spectral_features=np.hstack((spectral_features,SNR))
  return spectral_features
```

```python
def spectogram_formation(complex_data):
  fs= 1000
  spectrograms=[]
  for row in complex_data:
    nperseg = min(32, len(row))  # Ensure nperseg is not greater than the length of the row
    frequencies,time,sxx= spectrogram(row, fs, nperseg=nperseg)
    spectrograms.append(sxx)
  spectrograms=np.array(spectrograms)
  spectrogram_image_dB = 10 * np.log10(np.abs(spectrograms) + 1e-10)  # Adding a small value to avoid log of zero
  normalized_spectrogram_image_dB=(spectrogram_image_dB-np.min(spectrogram_image_dB))/(np.max(spectrogram_image_dB)-np.min(spectrogram_image_dB))
  normalized_spectrogram_image_dB= np.expand_dims(normalized_spectrogram_image_dB, axis=-1)
  print(normalized_spectrogram_image_dB.shape)
  return normalized_spectrogram_image_dB
```

```python
def data_cleaning(dataset, label, SNR):
  #dataset based
    clean_indice_dataset=~np.isnan(dataset).any(axis=(1,2)) & ~np.isinf(dataset).any(axis=(1,2))
    dataset=dataset[clean_indice_dataset]
    label=label[clean_indice_dataset]
    SNR=SNR[clean_indice_dataset]
    #label based
    clean_indice_label=~np.isnan(label).any(axis=1) & ~np.isinf(label).any(axis=1)
    label=label[clean_indice_label]
    dataset=dataset[clean_indice_label]
    SNR=SNR[clean_indice_label]
    #SNR based
    clean_indice_SNR=~np.isnan(SNR).any(axis=1) & ~np.isinf(label).any(axis=1)
    label=label[clean_indice_SNR]
    dataset=dataset[clean_indice_SNR]
    SNR=SNR[clean_indice_SNR]
    return dataset, label, SNR
```

```python
def complex_data(inphase_signal,quadrature_signal):
  complex_data= inphase_signal + 1j*quadrature_signal
  return complex_data
```
